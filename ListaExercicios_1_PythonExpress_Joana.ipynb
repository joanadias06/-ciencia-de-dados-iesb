```python
# 1) Elabore um código que imprima a mensagem *"Olá terráqueos!!"* na tela.

print("Olá terráqueos")
```

```python
# 2) Elabore um código que peça um número e então mostre a mensagem:
#"O número informado foi [número]."

x = int(input("Digite um número: "))
print("O número informado foi : ", x)
```

```python
# 3) Elabore um código que peça dois números e imprima a soma.
x = int(input("Digite um número: "))
y = int(input("Digite outro número: "))
print("A Soma dos dois números é de: ", x + y )
```

```python
# 4) Elabore um código que peça as 4 notas bimestrais e mostre a média.
nota_1 = float(input("Digite sua nota do primeiro bimestre: "))
nota_2 = float(input("Digite sua nota do segundo bimestre: "))
nota_3 = float(input("Digite sua nota do terceiro bimestre: "))
nota_4 = float(input("Digite sua nota do quarto bimestre: "))
media = (nota_1 + nota_2 + nota_3 + nota_4) / 4

print("A média das notas é de: ", media)
```

```python
# 5) Elabore um código que converta metros para centímetros.

x = float(input("Digite um número em metros que você deseja converter para centímetros:"))
operacao = x*100
print("O número convertido é: ", operacao)
```

```python
# 6) Elabore um código que peça o raio de um círculo, calcule e mostre sua área.
import math
raio = float(input("Digite o raio do círulo: "))
area = math.pi * (raio**2)
print("A área do círculo é de: ", area)
```

```python
# 7) Elabore um código que calcule a área de um quadrado, em seguida
#mostre o dobro desta área para o usuário.

lado = float(input("Digite um número: "))
area = lado**2
dobro = area*2
print("A Área do quadro é: ", area)
print("O dobro da Área é: ", dobro)
```

```python
# 8) Elabore um código que pergunte quanto você ganha por hora e
# o número de horas trabalhadas no mês. Calcule e mostre o total do seu salário
# no referido mês em uma mensagem amigável.

valor_hr = float(input("Digite quanto você ganha por hora: "))
hr_mes = float(input("Digite quantas horas você trabalha por mês: "))
salario = valor_hr * hr_mes
salario = salario * 30
print("Olá, tudo bem? o seu sálario no mês com 30 dias é de: ", salario)
```

```python
# 9) Elabore um código que peça a temperatura em graus Fahrenheit,
# transforme e mostre a temperatura em graus Celsius. ** Onde C = 5 * ((F-32) / 9)**.

fahrenheint = float(input("Digite a temperatura deseja em Fahrenheit:"))
celsius = 5 * ((fahrenheint - 32) / 9)
print("A temperatura em graus celsius é: ", celsius)
```

```python
# 10) Elabore um código que peça a temperatura em graus Celsius, transforme
#e mostre em graus Fahrenheit.

celsius = float(input("Digite a temperatura deseja em Celsius:"))
fahrenheint = (celsius * 9 / 5) + 32
print("A temperatura em fahrenheint é: ", fahrenheint)
```

```python
# 11) Elabore um código que peça 2 números inteiros e um número real. Calcule e mostre:
n1 = int(input("Digite um número inteiro: "))
n2 = int(input("Digite um número inteiro: "))
n3 = float(input("Digite um número real: "))

# a) o produto do dobro do primeiro com metade do segundo .
dobro = n1 * 2
metade = n2 / 2
produto = dobro * metade
print("O produto do dobro do primeiro com metade do segundo é de: ", produto)

# b) a soma do triplo do primeiro com o terceiro.
triplo = n1 * 3
soma = triplo + n3
print("A soma do triplo do primeiro com o terceiro é de: ", soma)

# c)o terceiro elevado ao cubo.
cubo = n3**3
print("O terceiro número elevado ao cubo é: ", cubo)
```

```python
# 12) Elabore um código que receba os dados de entrada a altura
#de uma pessoa, e informe qual seria o peso ideal,
#usando a seguinte fórmula: (72.7*altura) - 58

altura = float(input("Informe qual é a sua altura: "))
peso_ideal = (72.7 * altura) - 58

print("O seu peso ideal, conforme a sua altura é: ", peso_ideal)
```
